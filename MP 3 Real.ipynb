# Feature Engineering

## Membuat kolom umur customer

```python
 df['Year_Birth']=pd.to_datetime(df['Year_Birth'], format = '%Y')
```

```python
 df['Year_Birth']
```

```python
 df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'])
```

```python
df.info()
```

Membuat kolom umur (Customer_Age)

```python
df['Customer_Age']= np.abs(df['Dt_Customer']-df['Year_Birth'])
```

```python
df['Customer_Age']
```

```python
df['Customer_Age_Year']= df['Customer_Age']/np.timedelta64(1, 'Y')
```

```python
df[['Customer_Age_Year','Customer_Age']]
```

```python
df['Customer_Age_Year']=df['Customer_Age_Year'].apply(np.floor)
```

```python
df['Customer_Age_Year'].value_counts().sort_index()
```

## Membuat kelompok umur

```python
df['Customer_Age_Group']= np.where((df['Customer_Age_Year'] >= 0) & (df['Customer_Age_Year'] < 15),'Children',
                             np.where((df['Customer_Age_Year'] >= 15) & (df['Customer_Age_Year'] < 25),'Youth',
                                      np.where((df['Customer_Age_Year'] >= 25) & (df['Customer_Age_Year'] < 45),'Young Adults',                                               
                                               np.where((df['Customer_Age_Year'] >= 45) & (df['Customer_Age_Year'] < 65),'Middle-aged Adults','Old-aged Adults'))))
```

```python
df['Customer_Age_Group'].value_counts()
```

## Membuat kolom total anak

```python
df['Total_Kids']=df['Kidhome']+df['Teenhome']
```

```python
df['Total_Kids'].value_counts()
```

## Membuat kolom is_parents? (sudah mempunyai anak atau belum)

```python
df['Is_Parents']=np.where((df['Kidhome'] >= 1) & (df['Teenhome'] >= 1),'Yes','No')
```

```python
df['Is_Parents'].value_counts()
```

## Membuat kolom total pembelian 

```python
df['Total_Purchases']=df['NumDealsPurchases']+df['NumWebPurchases']+df['NumCatalogPurchases']
```

```python
df['Total_Purchases']
```

## Membuat kolom total pengeluaran

```python
df['Total_Outcomes']=df['MntCoke']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds']
```

```python
df['Total_Outcomes']
```

## Membuat kolom Conversion rate

```python
df['Conversion_Rate']=df['Response']/df['NumWebVisitsMonth']*100
```

```python
df['Conversion_Rate'].value_counts()
```

```python
df.info()
```

# Clustering

## Elbow Method

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import silhouette_score
```

```python
sse = {}
cr_data = df_clean2[['Recency_std','Total_Purchases_std','Total_Outcomes_std']]
for k in range(2,11):
  kmeans = KMeans(n_clusters = k, max_iter= 100, random_state=2022).fit(cr_data)
  cr_data['cluster'] = kmeans.labels_
  sse[k] = kmeans.inertia_
plt.figure(figsize=(19,6))
plt.plot(list(sse.keys()),list(sse.values()))
plt.xlabel('Number of Cluster')
```

```python
kmeans = KMeans(n_clusters = 4, max_iter= 100, random_state=2022).fit(cr_data)
cr_data['cluster'] = kmeans.labels_
```

```python
cr_data
```

```python
real_value=pd.concat([cr_data,df_copy], axis=1)
real_value
```

```python
real_value_viz=real_value[['cluster','Recency','Total_Purchases','Total_Outcomes']]
```

```python

fig = plt.figure(figsize = (5, 10))
ax = plt.axes(projection ="3d")
   

ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.3,
        alpha = 0.2)
 

x=real_value_viz['Recency']
y=real_value_viz['Total_Purchases']
z=real_value_viz['Total_Outcomes']
 

my_cmap = plt.get_cmap('Spectral')
 

sctt = ax.scatter3D(x, y, z,
                    alpha = 0.8,
                    c = real_value_viz['cluster'],
                    cmap = my_cmap,
                    )
 
plt.title("simple 3D scatter plot")
ax.set_xlabel('Recency', fontweight ='bold')
ax.set_ylabel('Total_Purchases', fontweight ='bold')
ax.set_zlabel('Total_Outcomes', fontweight ='bold')

 

plt.show()
```

```python
s_score = []

number_of_cluster = range(2,11)

for i in number_of_cluster:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(cr_data)
    labels = kmeans.labels_
    s_score.append(silhouette_score(cr_data, labels, metric = 'euclidean'))
plt.figure(figsize=(18,6))
sns.lineplot(number_of_cluster, s_score)
sns.scatterplot(number_of_cluster, s_score)
plt.xticks(number_of_cluster)
plt.xlabel('number of clusters')
plt.ylabel('silhouette score')
```
